```r
library(repr)
library(tidyverse)
library(tidymodels)


grades <-read_csv("StudentsPerformance.csv")

colnames(grades) <- make.names(colnames(grades))

grades_selected <- grades %>%
    select (gender, math.score, reading.score, writing.score)

grades_new <- grades_selected %>%
    mutate (average.score = (math.score + reading.score + writing.score) /3) %>%
    mutate (average.score = as_factor (average.score))

grades_new
```
